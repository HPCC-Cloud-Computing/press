```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Dataset
from sklearn.datasets import load_boston

boston = load_boston()
print(boston.data.shape)
print(type(boston))
```

```python
boston.data
```

```python
X = boston.data[:,:12]
y = boston.data[:,12]
```

```python
train_size = int(len(y) * 0.8)

# Train data
X_train = X[:train_size]
y_train = y[:train_size]

# Test data
X_test = X[train_size:]
y_test = y[train_size:]
```

```python
# Random forest model 
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(X_train, y_train)

# Predict
y_pred_rf = rf.predict(X_test)
```

```python
# Linear regression model
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)

# Predict
y_pred_lr = lr.predict(X_test)
```

```python
# Feed forward
from keras import Sequential
from keras.layers import Dense

model = Sequential()
model.add(Dense(units=100, activation='sigmoid', input_dim=(12)))
model.add(Dense(units=100, activation='sigmoid', input_dim=(12)))
model.add(Dense(units=1))
model.compile(loss='mse', optimizer='sgd')
model.fit(X_train, y_train, epochs=2000)

y_pred_ff = model.predict(X_test)
y_pred_ff = y_pred_ff.reshape(-1)
```

```python
# Error
from sklearn.metrics import mean_absolute_error, mean_squared_error

error = [mean_absolute_error, mean_squared_error]
for er in error:
    print(er.__name__ + ':')
    print('\tRandom forest:', er(y_pred_rf, y_test))
    print('\tLinear regression:', er(y_pred_lr, y_test))
    print('\tFeed forward:', er(y_pred_ff, y_test))
```
